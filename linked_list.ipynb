### Singly Linked List

```python
class Node:

    #constructor
    def __init__(self, data):

        self.data = data
        self.next = None

class topic:

    def arr_to_ll(self, arr):

        if len(arr) == 0:
            return None
            
        head = Node(arr[0])
        
        temp = head
        
        for i in range(1, len(arr)):
            
            newNode = Node(arr[i])
            temp.next = newNode
            temp = newNode
        
        return head
    
    def print_ll(self, head):

        temp = head
        while temp is not None:

            print(temp.data , end = "->")
            temp =  temp.next

        print("None")
        


arr = [1,2,3,4,5,6,7,8,9,0]

obj = topic()

head = obj.arr_to_ll(arr)

obj.print_ll(head)

```

### Doubly Linked List

```python
class Node:
    # constructor
    def __init__(self, data):

        self.data = data
        self.next = None
        self.prev = None


class Solution:

    def arr_to_dll(self, arr):

        n = len(arr)

        if n == 0:
            return None

        head = Node(arr[0])
        temp = head

        for i in range(1, n):

            newNode = Node(arr[i])

            temp.next = newNode
            newNode.prev = temp

            temp = newNode

        return head
    
    def insertAtPos(self, head, p, x):
        # Code Here
        
        newNode = Node(x)
        temp = head
        
        for i in range(p):
            temp = temp.next
            
        newNode.next = temp.next
        
        if temp.next:
            temp.next.prev = newNode
        
        temp.next = newNode
        newNode.prev = temp
    
        return head
    
    def delPos(self, head, x):
        # code here
        
        if x==1:
            head = head.next
            if head:
                head.prev = None
            return head
                
                
        temp = head
        
        
        for i in range(x-2):
            temp = temp.next
            
            
        nodeToDelete  = temp.next
        temp.next = nodeToDelete.next
        
        if nodeToDelete.next:
            nodeToDelete.next.prev = temp
        
        
        return head
    
    def reverse(self, head):
        # code here
        #for empty and single LL 
        if not head or not head.next:
            return head
            
        curr = head
        
        while curr:
            #swap prev and next node of LL
            curr.prev, curr.next = curr.next, curr.prev
            #shift to next
            head = curr
            curr = curr.prev
            
        return head

    def print_dll(self, head):

        temp = head

        while temp is not None:

            print(temp.data, end=" <-> ")
            temp = temp.next

        print("None")


arr = [1,2,3,4,5,6,7,8,9,0]

obj = Solution()

head = obj.arr_to_dll(arr)
# insert = obj.insertAtPos(head, 2, 11)
# delete = obj.delPos(head, 4)
rev = obj.reverse(head)

# obj.print_dll(head)
# obj.print_dll(insert)
# obj.print_dll(delete)
obj.print_dll(rev)
```
