##### Creation of Binary Tree


```python
# Creating Binary Tree

class TreeNode:
    def __init__ (self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
root = TreeNode('R')
nodeA = TreeNode('A')
nodeB = TreeNode('B')
nodeC = TreeNode('C')
nodeD = TreeNode('D')
nodeE = TreeNode('E')
nodeF = TreeNode('F')
nodeG = TreeNode('G')

root.left = nodeA
root.right = nodeB

nodeA.left = nodeC
nodeA.right = nodeD

nodeB.left = nodeE
nodeB.right = nodeF

nodeF.left = nodeG

# Tree Structure:
#           R
#         /   \
#        A     B
#       / \   / \
#      C   D E   F
#               /
#              G

# Test
print(root.left.right.data)
```

### Traversals in Binary Tree

##### Inorder (Left → Root → Right)

```python
# Recursive Methord

def inorder(root):
    if root is None:
        return
    inorder(root.left)
    print(root.data, end=" ")
    inorder(root.right)
```

##### Preorder (Root → Left → Right)

```python
# Recursive Methord

def preorder(root):
    if root is None:
        return
    print(root.data, end=" ")
    preorder(root.left)
    preorder(root.right)
```

##### Postorder (Left → Right → Root)

```python
# Recursive Methord

def postorder(root):
    if root is None:
        return
    postorder(root.left)
    postorder(root.right)
    print(root.data, end=" ")
```

## BFS

```python
from collections import deque

def bfs(node):
    if node is None:
        return
    queue = deque()
    queue.append(node) # Start with root

    while queue:
        current = queue.popleft()  # Remove form front
        print(current.data, end=" ")

        if current.left:           # Add left child
            queue.append(current.left) 

        if current.right:           # Add right child
            queue.append(current.right)
```

##### Inorder (Root → Left → Right)

$Algorithm : $

Iterative Methord

1. Keep going LEFT and push nodes.
2. When LEFT ends, POP and process.
3. Move to RIGHT.
4. Repeat.

```python
# Iterative

def inorder_iterative(root):
    stack = []
    curr = root

    while curr or stack:
        while curr:
            stack.append(curr)
            curr = curr.left
        
        curr = stack.pop()
        print(curr.data, end=" ")

        curr = curr.right
```

##### Preorder (Root → Left → Right)

$Algorithm : $

Iterative Methord

1. Push root into `Stack` #LIFO

2. While stach is not empty :

    - Pop top node.
    - Print/Process it.
    - Push its right child first
    - Push its left child second

Since stack is LIFO, the left child will be processed before the right child.

```python
#Iterative Methord

def preorder_iterative(root):

    if not root:
        return
    
    stack = [root]
    ans = []

    while stack:

        node = stack.pop()
        print(node.data, end = " ")

        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
```

##### Postorder (Left → Right → Root)

$Algorithm : $

Iterative Methord

1. Push root into `stack1`.
2. Pop from `stack1` and push into `stack2`.
3. Push left and right childern to  `stack1`.
4. Repeat till `stack1` empty.
5. Pop all elements form `stack2`.

```python
def postorder_iterative(root):

    if not root:
        return

    s1 = [root]
    s2 = []

    while s1:
        node = s1.pop()
        s2.append(node)

        if node.left:
            s1.append(node.left)
        if node.right:
            s1.append(node.right)
    
    while s2:
        node = s2.pop()
        print(node.data, end=" ")
```
